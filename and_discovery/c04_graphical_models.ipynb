# Graphs

- Undirected vs directed
- Cyclic vs acyclic
- Connected versus disconnected
- Weighted vs unweighted

```python
import networkx as nx

sample_gml = """graph[
    directed 1

    node [
        id 0
        label "0"
    ]

    node [
        id 1
        label "1"
    ]

    node [
        id 2
        label "2"

    ]

    edge [
        source 0
        target 1
    ]

    edge [
        source 2
        target 1
    ]
]
"""

graph = nx.parse_gml(sample_gml)

nx.draw(graph,
        with_labels=True,
        font_color="white",
        node_size=2000)
```
