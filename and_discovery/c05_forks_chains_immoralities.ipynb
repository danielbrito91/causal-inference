```python
import numpy as np
import graphviz as gr
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
```

```python
NOISE_LEVEL = .2
N_SAMPLES = 1000
```

# Chain

```python
# CHAIN
g = gr.Digraph()
g.edge("A", "B")
g.edge("B", "C")

a = np.random.randn(N_SAMPLES)
b = a + NOISE_LEVEL * np.random.randn(N_SAMPLES)
c = b + NOISE_LEVEL * np.random.randn(N_SAMPLES)

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(a, b, alpha=.3)
axs[0, 1].scatter(c, a, alpha=.3)
axs[1, 0].scatter(b, c, alpha=.3)

plt.suptitle("Chain")
plt.show()

g
```

```python
X = pd.DataFrame(np.vstack([a, b]).T, columns=["A", "B"])
X = sm.add_constant(X)
model = sm.OLS(c, X)
results = model.fit()
results.summary().tables[1]
```

# Fork

```python
b = np.random.randn(N_SAMPLES)
a = b + NOISE_LEVEL * np.random.randn(N_SAMPLES)
c = b + NOISE_LEVEL * np.random.randn(N_SAMPLES)

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(a, b, alpha=.3)
axs[0, 1].scatter(c, a, alpha=.3)
axs[1, 0].scatter(b, c, alpha=.3)

plt.suptitle("Fork")
plt.show()
```

```python
X = pd.DataFrame(np.vstack([a, b]).T, columns=["A", "B"])
X = sm.add_constant(X)
model = sm.OLS(c, X)
results = model.fit()
results.summary().tables[1]
```

# Collider

```python
a = np.random.randn(N_SAMPLES)
c = np.random.randn(N_SAMPLES)
b = a + c + NOISE_LEVEL * np.random.randn(N_SAMPLES)

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(a, b, alpha=.3)
axs[0, 1].scatter(c, a, alpha=.3)
axs[1, 0].scatter(b, c, alpha=.3)

plt.suptitle("Collider")
plt.show()
```

```python
X = pd.DataFrame(np.vstack([a, b]).T, columns=["A", "B"])
X = sm.add_constant(X)
model = sm.OLS(c, X)
results = model.fit()
results.summary().tables[1]
```
